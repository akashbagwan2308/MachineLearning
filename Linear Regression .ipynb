# Linear Regression 

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets,linear_model
from sklearn.metrics import mean_squared_error
```

```python
diabetes = datasets.load_diabetes()
```

```python
# ['data', 'target', 'frame', 'DESCR', 'feature_names', 'data_filename', 'target_filename', 'data_module']
#print(diabetes.keys())
```

```python
diabetes_X = diabetes.data[:,np.newaxis,2]
# diabetes_X = diabetes.data
# print(diabetes_X)
```

**Dividing dataset into train and test data** 

```python
diabetes_X_train = diabetes_X[:-30]
diabetes_X_test = diabetes_X[-30 : ]

diabetes_Y_train = diabetes.target[:-30]
diabetes_Y_test = diabetes.target[-30 : ]
```

**Making Linear Regression Model**

```python
model = linear_model.LinearRegression()
model.fit(diabetes_X_train,diabetes_Y_train)
```

**Making Prediction**

```python
diabetes_Y_predict = model.predict(diabetes_X_test )

print(f"Mean squared error : {mean_squared_error(diabetes_Y_test,diabetes_Y_predict)}")
```

**Getting Weights and Intercept**

```python
print(f"Weights : {model.coef_}\nIntercept : {model.intercept_}" )
```

**PLoting Graph**

```python
plt.scatter(diabetes_X_test,diabetes_Y_test,c="r")
plt.plot(diabetes_X_test,diabetes_Y_predict)
plt.xlabel("X_test_data")
plt.ylabel("Y_predicted")

plt.show()
```
